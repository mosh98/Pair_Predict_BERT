```python
from transformers import AutoModelForSequenceClassification,AutoTokenizer
import torch
device = torch.device("cuda" if torch.cuda.is_available() else 'cpu')

model = AutoModelForSequenceClassification.from_pretrained("siebert/sentiment-roberta-large-english")
tokenizer = AutoTokenizer.from_pretrained("siebert/sentiment-roberta-large-english",truncation=True,return_tensors="pt")
```

#Define Text


```python
text_A = "There is a bunger joint in stockholm"
text_B = "The burgers tasted like crap"
```

#Step 1


```python
tokenized_items = tokenizer(text_A, text_B, return_tensors="pt",truncation=True)
```

```python
tokenized_items.keys()
```

#Step 2



```python
import torch
device = torch.device("cuda" if torch.cuda.is_available() else 'cpu')
```

```python
  
tokenized_items.to(device) #set tokenized_item to device
model.to(device) #set model to device
  
```

#Step 3



```python
#get output
output_logits = model(**tokenized_items).logits
```

```python
output_logits
```

#Step 4



```python
output_results = torch.softmax(output_logits, dim=1).tolist()[0]
```

```python
output_results
```

#Step 5



```python
#process output
maxVal = max(output_results)
  
#print(maxVal)
index = output_results.index(maxVal) #idex of max value
  
#Conversion 
dict_items = model.config.id2label.items()
myList = [(k,v) for k,v in model.config.id2label.items()] #id2Label

predictedLabel = myList[index]
```
